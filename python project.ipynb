```python
#drop blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#handling null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
#EDA(EXPLORATORY DATA ANALYSIS)
```

```python
df.columns
```

```python
#GENDER
```

```python
#GENDER COUNT
a= sns.countplot(x = 'Gender',data = df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
#GENDER VS AMOUNT
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
#Females make more purchases than males
```

```python
#AGE
```

```python
a = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
#AGE GROUP VS AMOUNT
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
#Age group between 26-35 make more purchases
```

```python
#STATE
```

```python
#ORDERS FROM TOP TEN STATES
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
#SALES STATE VS AMOUNT
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
#More orders are from Uttar pradesh ,maharashtra and karnataka
```

```python
#MARITAL STATUS
```

```python
a = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(20,5)})
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
#Married people make more purchases
```

```python
#OCCUPATION
```

```python
a = sns.countplot(data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
#People in IT sector,Healthcare and Aviation makes more purchases
```

```python
#PRODUCT CATEGORY
```

```python
a = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
#Most Sold products are from Food,Clothing and Electronics category
```

```python
#FROM ABOVE ANALYSIS WE CAN CONCLUDE THAT MOST SOLD PRODUCTS ARE FROM FOOD,CLOTHING AND ELECTRONICS CATEGORY AND THE BUYERS ARE MARRIED WOMEN BETWEEN AGE GROUP 26-35 WORKING IN IT SECTOR,HEALTHCARE AND AVIATION AND ARE FROM STATES(UTTAR PRADESH,MAHARASHTRA AND KARNATAKA)
```
